```python
import os
import pydicom

root_dir = r'D:\Code\Tuharririna\data\trainer\ddsm\images\dicom\manifest-ZkhPvrLo5216730872708713142\CBIS-DDSM'

count = 0
max_images = 30

for root, dirs, files in os.walk(root_dir):
    for file in files:
        if file.lower().endswith('.dcm'):
            dcm_path = os.path.join(root, file)

            try:
                ds = pydicom.dcmread(dcm_path)
                image = ds.pixel_array

                height, width = image.shape

                print(f"{count+1}. {file}")
                print(f"   Width : {width}")
                print(f"   Height: {height}")
                print(f"   Path  : {dcm_path}\n")

                count += 1
                if count >= max_images:
                    break

            except Exception as e:
                print(f"Skipping {dcm_path} (error reading DICOM)")
                print(e)

    if count >= max_images:
        break
```
